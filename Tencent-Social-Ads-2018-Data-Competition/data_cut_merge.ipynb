```python
import pandas as pd
import numpy as np
import time
import warnings
import re
import glob    #用来读取文件夹中的所有文件
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 50)
```

```python
adFeature = pd.read_csv('preliminary_contest_data/adFeature.csv')
```

```python
##--把训练集标签-1转为0--##
train = pd.read_csv('preliminary_contest_data/train.csv')    
### train.ix[train['label']==-1, 'label'] = 0
### train.to_csv('preliminary_contest_data/train.csv', index=False)
```

```python
##--按aid切分train，并单独保存--##
train_groups = train.groupby('aid')
count = 0;
for aid, group in train_groups:
    group.to_csv('./cut_train/aid_%s.csv'%aid, index=False);
    count = count + 1;
    print('aid=%s, %d groups saved.'%(aid,count))
```

```python
##--将train按aid切分出来的文件，与adFeature按aid进行左外连接--##
start = time.clock()
import glob
file_list = glob.glob('./cut_train/*.csv')
count = 0;
for file in file_list:
    aid = re.split('\.|/|_|',file)[-2]   ##--由文件路径，获取aid编号--##
    cur_df = pd.read_csv(file);
    cur_df = cur_df.merge(adFeature, on='aid', how='left');
    cur_df.to_csv('./merge_ad/merge_ad_%s.csv'%aid, index=False);
    count = count+1;
    time_cost = time.clock()-start;
    print('aid=%s, %d groups merged. %.2f s cost.'%(aid, count, time_cost))
```

```python
userFeature = pd.read_csv('preliminary_contest_data/userFeature.csv')
```

```python
##-----将上面的连接结果再与userFeature进行左外连接----##
start = time.clock()
import glob
file_list = glob.glob('./merge_ad/*.csv')
count = 0;
for file in file_list:
    cur_start = time.clock();
    aid = re.split('\.|/|_|',file)[-2]
    cur_df = pd.read_csv(file);
    cur_df = cur_df.merge(userFeature, on='uid', how='left');
    cur_df.to_csv('./merge_user/merge_user_%s.csv'%aid, index=False);
    count = count + 1;
    time_cost = time.clock() - start; cur_cost= time.clock() - cur_start; m = int(time_cost / 60);  s = time_cost % 60;
    rtc = (173-count)*cur_cost; rtm = int(rtc / 60); rts = rtc % 60;
    print('aid=%s, %d groups merged. %d m %d s cost. last group cost %d s. rtc: %d m %d s.'%(aid, count, m, s, cur_cost, rtm, rts))
```
